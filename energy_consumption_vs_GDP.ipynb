```python
import warnings
warnings.filterwarnings('ignore')

# data imports
import numpy as np
from plotnine import *
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#world energy data set
energy_world = pd.read_csv("/content/drive/MyDrive/Projects/owid-energy-data.csv")
```

```python
#selecting and cleaning the dataset
selected_columns = ['year', 'country', 'population', 'gdp', 'energy_per_capita']
energy_gdp = energy_world[selected_columns]
countries = ['China', 'India', 'United States']
years = [2000, 2005, 2010, 2015, 2020]
energy_gdp_df = energy_gdp[(energy_gdp['country'].isin(countries)) & (energy_gdp['year'].isin(years))]
energy_gdp_df['gdp_per_capita'] = energy_gdp_df['gdp'] / energy_gdp_df['population']
print(energy_gdp_df)
```

```python

# Assuming the dataframe is stored in 'df'
# Filter the data for the relevant countries

# Create a scatter plot with total GDP vs energy consumption per capita for each country
plt.figure(figsize=(10, 6))
sns.lineplot(data=energy_gdp_df, x='gdp', y='energy_per_capita', hue='country', style='country', palette='Set1')

# Add labels and title
plt.title('GDP vs Energy Consumption per Capita')
plt.xlabel('GDP (USD)')
plt.ylabel('Energy Consumption per Capita (kWh)')

# Show the plot
plt.legend(title='Country')
plt.show()
```
