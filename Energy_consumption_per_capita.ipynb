```python
import warnings
warnings.filterwarnings('ignore')

# data imports
import pandas as pd
import numpy as np
from plotnine import *

# modeling imports
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#world energy data set
energy_world = pd.read_csv("/content/drive/MyDrive/Projects/owid-energy-data.csv")
```

```python
#cleaning and selecting columns for new data frame
country_energy_columns = ['year', 'country', 'energy_per_capita']
country_energy_consumption = energy_world[country_energy_columns]
countries = ['China', 'India', 'United States']
years = [2000, 2005, 2010, 2015, 2020]
country_energy_consumption_df = country_energy_consumption[(country_energy_consumption['country'].isin(countries)) & (country_energy_consumption['year'].isin(years))]
print(country_energy_consumption_df)
```

```python

# Filter data by country
countries = country_energy_consumption_df['country'].unique()

# Plot each country as a separate line
for country in countries:
    country_data = country_energy_consumption_df[country_energy_consumption_df['country'] == country]
    plt.plot(country_data['year'], country_data['energy_per_capita'], label=country)

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Energy Consumption Per Capita (Terrawatt-hour/person)')
plt.title('Energy Consumption per Capita Over Time by Country')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
