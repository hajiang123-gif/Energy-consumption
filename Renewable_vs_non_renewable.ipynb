```python
#world energy data set
energy_world = pd.read_csv("/content/drive/MyDrive/Projects/owid-energy-data.csv")
```

```python
#selecting and cleaning data from data frame
selected_columns = ['year', 'country', 'wind_consumption', 'solar_consumption', 'nuclear_consumption', 'hydro_consumption', 'biofuel_consumption', 'coal_consumption', 'fossil_fuel_consumption', 'gas_consumption', 'oil_consumption']
energy_sectoral_country = energy_world[selected_columns]
countries = ['China', 'India', 'United States']
years = [2000, 2020]
energy_sectoral_country_df = energy_sectoral_country[(energy_sectoral_country['country'].isin(countries)) & (energy_sectoral_country['year'].isin(years))]
print(energy_sectoral_country_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = energy_sectoral_country_df

# Filter the data for China in 2000
china_2000 = df[(df["country"] == "China") & (df["year"] == 2000)]

# Extract energy types and consumption values
energy_types = [
    "wind_consumption",
    "solar_consumption",
    "nuclear_consumption",
    "hydro_consumption",
    "biofuel_consumption",
    "coal_consumption",
    "fossil_fuel_consumption",
    "gas_consumption",
    "oil_consumption",
]
consumption_values = china_2000[energy_types].iloc[0]

# Calculate percentages
total_consumption = consumption_values.sum()
percentages = (consumption_values / total_consumption) * 100

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    percentages,
    labels=energy_types,
    autopct="%1.1f%%",
    startangle=140,
    colors=plt.cm.tab10.colors,
)
plt.title("Energy Consumption Distribution in China (Year 2000)")
plt.tight_layout()

# Show plot
plt.show()


```
